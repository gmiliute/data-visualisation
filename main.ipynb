```python
onesky_data['reading_time'] = pd.to_datetime(onesky_data['reading_time'])

def convert_to_decimal(coord_str):
    # Check if the string is empty or not in the expected format
    if not coord_str or not coord_str[:-1].isdigit():
        return None  # Return None or an appropriate value for invalid data
    
    # Extract the numeric part and the direction
    numeric_part = float(coord_str[:-1]) / 1000000
    direction = coord_str[-1]
    
    # Adjust the sign based on the direction
    if direction in ['S', 'W']:
        numeric_part *= -1
    return numeric_part

# Apply conversion to latitude and longitude with updated function
onesky_data['latitude'] = onesky_data['latitude'].apply(convert_to_decimal)
onesky_data['longitude'] = onesky_data['longitude'].apply(convert_to_decimal)

# Check the updated dtypes and head of the DataFrame
print(onesky_data.dtypes)
print(onesky_data.info())
```
